```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
# Review combined dataframe
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calculate the total number of schools
total_schools = school_data_complete["school_name"].nunique()

# Calculate the total number of students
total_students = school_data_complete["Student ID"].nunique()

# Calculate the total budget
total_budget = school_data["budget"].sum()

# Calculate the average math score 
ave_math = school_data_complete["math_score"].mean()

# Calculate the average reading score
ave_read = school_data_complete["reading_score"].mean()

# Calculate the percentage of students with a passing math score (70 or greater)
prct_pass_math = len(school_data_complete.loc[school_data_complete['math_score']>=70]) / total_students

# Calculate the percentage of students with a passing reading score (70 or greater)
prct_pass_read = len(school_data_complete.loc[school_data_complete['reading_score']>=70]) / total_students

# Calculate the percentage of students who passed math and reading (% Overall Passing)
pass_both = school_data_complete['student_name'].loc[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)].count() 
prct_pass_both = (pass_both / total_students)

# Create dataframe to house District Summary data
district_summary_df = pd.DataFrame({"Total Schools": [total_schools],
    "Total Students": [total_students],
    "Total Budget": [total_budget],
    "Average Math Score": [ave_math],
    "Average Reading Score": [ave_read],
    "% Passing Math": [prct_pass_math],
    "% Passing Reading": [prct_pass_read],
    "% Overall Passing": [prct_pass_both]})
columns = ["Total Schools","Total Students","Total Budget","Average Math Score","Average Reading Score",
"% Passing Math","% Passing Reading","% Overall Passing"]
    
# Format cells
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)
district_summary_df["Average Math Score"] = district_summary_df["Average Math Score"].map("{:.6f}".format)
district_summary_df["Average Reading Score"] = district_summary_df["Average Reading Score"].map("{:.6f}".format)
district_summary_df["% Passing Math"] = district_summary_df["% Passing Math"].map("{:.4%}".format)
district_summary_df["% Passing Reading"] = district_summary_df["% Passing Reading"].map("{:.4%}".format)
district_summary_df["% Overall Passing"] = district_summary_df["% Overall Passing"].map("{:.4%}".format)

district_summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Group data by unique school name 
grp_schools = school_data_complete.groupby(['school_name'])
school = school_data_complete['school_name'].unique()

# Show school type by school --commented is my original code.  Found set_index via Google research
# grp_type = school_data_complete.groupby(['school_name','type'])
# school_type = school_data_complete['type'].unique()
school_type = school_data.set_index('school_name')['type']

# Calculate the total number of students per school
students_per_school = grp_schools['Student ID'].count()

# Calculate the total budget per school
budget_per_school = school_data_complete['budget'].unique()

# Calculate the per student budget by school
size = school_data['size']
per_student_budget = budget_per_school / size

# Calculate the average math score by school 
ave_math_per_school = grp_schools['math_score'].mean()

# Calculate the average reading score by school
ave_read_per_school = grp_schools['reading_score'].mean()

# Calculate the percentage of students with a passing math score (70 or greater) by school
pass_math_byschool = school_data_complete.loc[school_data_complete['math_score']>=70].groupby('school_name')['math_score'].count()/students_per_school 
                                
# Calculate the percentage of students with a passing reading score (70 or greater) by school
pass_read_byschool = school_data_complete.loc[school_data_complete['reading_score']>=70].groupby('school_name')['reading_score'].count()/students_per_school 

# Calculate the percentage of students who passed math and reading (% Overall Passing) by school
pass_both_byschool = school_data_complete.loc[(school_data_complete['math_score'] >= 70) & (school_data_complete['reading_score'] >= 70)].groupby('school_name')['Student ID'].count()/students_per_school 
 
# Create dataframe to house School Summary data
school_summary = {
    "School Name": school,
    "School Type": school_type,
    "Total Students": students_per_school,
    "Total School Budget": budget_per_school,
    "Per Student Budget": per_student_budget,
    "Average Math Score": ave_math_per_school,
    "Average Reading Score": ave_read_per_school,
    "% Passing Math": pass_math_byschool,
    "% Passing Reading": pass_read_byschool,
    "Overall Passing Rate": pass_both_byschool}
# This section of code was used from Stackoverflow.com as I was running into an ValueError where arrays must all be the same length.
school_summ_df = pd.DataFrame.from_dict(school_summary, orient='index')
school_summ_df = school_summ_df.transpose()
    
# Format cells
school_summ_df["Total School Budget"] = school_summ_df["Total School Budget"].map("${:,.2f}".format)
school_summ_df["Average Math Score"] = school_summ_df["Average Math Score"].map("{:.6f}".format)
school_summ_df["Average Reading Score"] = school_summ_df["Average Reading Score"].map("{:.6f}".format)
school_summ_df["% Passing Math"] = school_summ_df["% Passing Math"].map("{:.2%}".format)
school_summ_df["% Passing Reading"] = school_summ_df["% Passing Reading"].map("{:.2%}".format)
school_summ_df["Overall Passing Rate"] = school_summ_df["Overall Passing Rate"].map("{:.2%}".format)

school_summ_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Create bins and labels for new dataframe
spend_bins = [0, 599, 630, 645, 700]
group_names = ["<$600", "$600-630", "$630-645", "$645-700"]
school_data_complete["spending_range"] = pd.cut(school_data_complete['budget']/school_data_complete['size'], spend_bins, labels=group_names)

# Group data by Spending
spend_group = school_data_complete.groupby('spending_range')

# Complete calculations on grouped data
math_avg = spend_group['math_score'].mean()
reading_avg = spend_group['reading_score'].mean()
math_pass = school_data_complete[school_data_complete['math_score']>=70].groupby('spending_range')['Student ID'].count()/spend_group['Student ID'].count()
read_pass = school_data_complete[school_data_complete['reading_score']>=70].groupby('spending_range')['Student ID'].count()/spend_group['Student ID'].count()
both_pass = school_data_complete.loc[(school_data_complete['math_score'] >= 70) & (school_data_complete['reading_score'] >= 70)].groupby('spending_range')['Student ID'].count()/spend_group['Student ID'].count()

# Create dataframe
spending_df = pd.DataFrame({
        "Average Math Score": math_avg,
        "Average Reading Score": reading_avg,
        "% Passing Math": math_pass,
        "% Passing Reading": read_pass,
        "% Overall Passing Rate": both_pass
    },
    columns=["Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing Rate"])

# Format cells
spending_df["% Passing Math"] = spending_df["% Passing Math"].map("{:.2%}".format)
spending_df["% Passing Reading"] = spending_df["% Passing Reading"].map("{:.2%}".format)
spending_df["% Overall Passing Rate"] = spending_df["% Overall Passing Rate"].map("{:.2%}".format)

spending_df
```

## Scores by School Size

```python
# Create bins and labels for new dataframe
size_bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
school_data_complete["school_size"] = pd.cut(school_data_complete['size'], size_bins, labels=group_names)

# Group data by school size
school_size = school_data_complete.groupby('school_size')

# Complete calculations on grouped data
math_avg = school_size['math_score'].mean()
reading_avg = school_size['reading_score'].mean()
math_pass = school_data_complete[school_data_complete['math_score']>=70].groupby('school_size')['Student ID'].count()/school_size['Student ID'].count()
read_pass = school_data_complete[school_data_complete['reading_score']>=70].groupby('school_size')['Student ID'].count()/school_size['Student ID'].count()
both_pass = school_data_complete.loc[(school_data_complete['math_score'] >= 70) & (school_data_complete['reading_score'] >= 70)].groupby('school_size')['Student ID'].count()/school_size['Student ID'].count()

# Create dataframe
size_df = pd.DataFrame({
        "Average Math Score": math_avg,
        "Average Reading Score": reading_avg,
        "% Passing Math": math_pass,
        "% Passing Reading": read_pass,
        "% Overall Passing Rate": both_pass
    },
    columns=["Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing Rate"])

# Format cells
size_df["% Passing Math"] = size_df["% Passing Math"].map("{:.2%}".format)
size_df["% Passing Reading"] = size_df["% Passing Reading"].map("{:.2%}".format)
size_df["% Overall Passing Rate"] = size_df["% Overall Passing Rate"].map("{:.2%}".format)

size_df
```
